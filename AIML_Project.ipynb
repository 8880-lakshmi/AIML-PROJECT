```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##### Student Score Card Analysis #####

```python
#load data set
df = pd.read_csv('/content/Student score card 1 (1).csv')
```

```python
print(df.head())
print(df.tail())
```

```python
score = np.array(df)
print(score)
```

```python
#calculate statistics
data ={
    'mean': df['Total'].mean(),
    'median':df['Total'].median(),
    'mode':df['Total'].mode(),
    'std':df['Total'].std()
}
```

```python
print(data)
```

```python
#plot histogram
plt.hist(df['Total'])
```

```python
#performance spread
mean_total=data['mean']
std_total=data['std']
performance=np.where(df['Total']>=(mean_total+std_total),"higher achiever","regular achiever")
print(performance)
```
